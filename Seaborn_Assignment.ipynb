Five plots that can be plotted using the Seaborn library and their uses:

Scatter Plot: Used to visualize the relationship between two continuous variables.
Line Plot: Used to show the trend or relationship between two variables over a continuous interval.
Bar Plot: Used to compare categorical variables and display their counts or aggregate values.
Box Plot: Used to display the distribution and statistical summary of numerical variables.
Violin Plot: Combines a box plot and a kernel density plot to show the distribution of a variable.

```python
import seaborn as sns

fmri_data = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
```

```python
import seaborn as sns

titanic_data = sns.load_dataset("titanic")
sns.boxplot(x="pclass", y="age", data=titanic_data)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
```

```python
import seaborn as sns

diamonds_data = sns.load_dataset("diamonds")
sns.histplot(data=diamonds_data, x="price", hue="cut")
```

```python
import seaborn as sns

iris_data = sns.load_dataset("iris")
sns.pairplot(data=iris_data, hue="species")
```

```python
import seaborn as sns

flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot("month", "year", "passengers")
sns.heatmap(data=flights_pivot, annot=True, cmap="YlGnBu")
```
